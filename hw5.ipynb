```python
def makeData ():
    # Строим сетку в интервале от -10 до 10 с шагом 0.1 по обоим координатам
    x2 = np.arange (-100, 500, 1)
    y2 = np.arange (-200, 500, 1)

    # Создаем двумерную матрицу-сетку
    xgrid, ygrid = np.meshgrid(x2, y2)

    # В узлах рассчитываем значение функции
    zgrid = (xgrid**2) + (ygrid**2)
    return xgrid, ygrid, zgrid
```

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D

x2, y2, z2 = makeData()

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x2, y2, z2, rstride=5, cstride=5, color='#11aa55')
axes.plot([0, 200], [0, 400], [0, 100**2+200**2])
pylab.show()
```
